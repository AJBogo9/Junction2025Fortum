```python
# %% Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# %% Load the data
monthly_forecast = pd.read_csv('data/fva_data/monthly_forecast.csv', sep=';', decimal=',', parse_dates=['measured_at'])
validation_actual = pd.read_csv('data/fva_data/validation_12m_actual.csv', sep=';', decimal=',', parse_dates=['measured_at'])
validation_baseline = pd.read_csv('data/fva_data/validation_12m_baseline.csv', sep=';', decimal=',', parse_dates=['measured_at'])
```

```python
# Get list of customer groups (all columns except 'measured_at')
customer_groups = [col for col in monthly_forecast.columns if col != 'measured_at']

print(f"Available customer groups: {customer_groups}")
```

```python
# %% Choose customer group to visualize
# CHANGE THIS TO SELECT DIFFERENT CUSTOMER GROUP
selected_group = '39'  # Change this value to any customer group ID

# %% Plot time series for selected customer group
fig, ax = plt.subplots(figsize=(12, 6))

# # Plot forecast
# ax.plot(monthly_forecast['measured_at'], 
#         monthly_forecast[selected_group], 
#         label='Forecast', 
#         marker='o', 
#         linewidth=2)

# Plot actual
ax.plot(validation_actual['measured_at'], 
        validation_actual[selected_group], 
        label='Actual', 
        marker='s', 
        linewidth=2)

# # Plot baseline
# ax.plot(validation_baseline['measured_at'], 
#         validation_baseline[selected_group], 
#         label='Baseline', 
#         marker='^', 
#         linewidth=2,
#         alpha=0.7)

ax.set_title(f'Customer Group {selected_group} - Time Series', fontsize=14, fontweight='bold')
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Value', fontsize=12)
ax.legend(fontsize=11)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
