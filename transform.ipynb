```python
import pandas as pd
```

```python
excel_path = "data/20251111_JUNCTION_training.xlsx"

cons_df = pd.read_excel(excel_path, sheet_name="training_consumption")
price_df = pd.read_excel(excel_path, sheet_name="training_prices")
groups_df = pd.read_excel(excel_path, sheet_name="groups")
```

```python
id_cols = ["measured_at"]

cons_long = cons_df.melt(
    id_vars=id_cols,
    var_name="group_id",
    value_name="consumption_fwh"
)

# Ensure correct types
cons_long["measured_at"] = pd.to_datetime(cons_long["measured_at"])
cons_long["group_id"] = cons_long["group_id"].astype(str)
```

```python
price_df["measured_at"] = pd.to_datetime(price_df["measured_at"])

merged = cons_long.merge(price_df, on="measured_at", how="left")
```

```python
groups_df["group_id"] = groups_df["group_id"].astype(str)

merged = merged.merge(groups_df, on="group_id", how="left")

# Save full dataset
full_output_path = "data/merged_hourly_for_azure_full.csv"
merged.to_csv(full_output_path, index=False)
print(f"Saved full data to {full_output_path}")
```

```python
# Filter to exclude last 12 months (keep data up to 2023-09-30)
cutoff_date = pd.to_datetime('2023-10-01T00:00:00.000Z')
merged_filtered = merged[merged['measured_at'] < cutoff_date]

# Save filtered dataset
filtered_output_path = "data/merged_hourly_for_training_up_to_2023_sep.csv"
merged_filtered.to_csv(filtered_output_path, index=False)
print(f"Saved filtered data to {filtered_output_path}")
merged_filtered.head()
```
